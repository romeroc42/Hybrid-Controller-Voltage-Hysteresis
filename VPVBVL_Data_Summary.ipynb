```python
#Create a matrix to represent the x-axis of a graph
#x1 = np.arange(0, len(y), 1)
x1 = np.arange(0, len(data_mult["ID"]), 1)
x1
```

```python
y1 = data_mult[["Voltaje Panel"]]
```

```python
#Power vs Time
plt.figure(figsize=(24,10))
plt.title("Voltaje Panel vs Tiempo", fontsize = 16)
plt.scatter(x1, y1, s=5, color="red", label="Voltaje Panel")
plt.plot(x1, y1, lw=0.8, color="blue", label="Voltaje Panel")
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.xlabel('Tiempo', fontsize = 16)
plt.ylabel('Voltaje Panel', fontsize = 16)
plt.legend(fontsize = 16)
plt.show()
```

```python
#Show Voltage Variable
y2 = data_mult[["Voltaje Batería"]]
print(y2)
```

```python
#Voltage vs Time
plt.figure(figsize=(24,10))
plt.title("Voltaje Batería vs Tiempo", fontsize = 16)
plt.scatter(x1, y2, s=5, color="red", label="Voltaje Batería")
plt.plot(x1, y2, lw=0.8, color="blue", label="Voltaje Batería")
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.xlabel('Tiempo', fontsize = 16)
plt.ylabel('Voltaje Batería', fontsize = 16)
plt.legend(fontsize = 16)
plt.show()
```

```python
#Show Current Variable
y3 = data_mult[["Voltaje Carga"]]
print(y3)
```

```python
#Current vs Time
plt.figure(figsize=(24,10))
plt.title("Voltaje Carga vs Tiempo", fontsize = 16)
plt.scatter(x1, y3, s=5, color="red", label="Voltaje Carga")
plt.plot(x1, y3, lw=0.8, color="blue", label="Voltaje Carga")
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)
plt.xlabel('Tiempo', fontsize = 16)
plt.ylabel('Voltaje Carga', fontsize = 16)
plt.legend(fontsize = 16)
plt.show()
```
